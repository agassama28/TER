```python
nb_simulations = 1000
p = 3 / 5
taille_echantillon = [10, 100, 1000, 10000]
mu = p

fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for idx, n in enumerate(taille_echantillon):
    moyennes = []
    sigma = np.sqrt(p * (1 - p) / n)

    for i in range(nb_simulations):
        X = np.random.binomial(1, p, n)
        moyenne = np.mean(X)
        moyennes.append(moyenne)

    ax = axes[idx]
    ax.hist(moyennes, bins=30, density=True, alpha=0.6, color='blue', label='Histogramme')
    x = np.linspace(min(moyennes), max(moyennes), 200)
    ax.plot(x, norm.pdf(x, mu, sigma), lw=2, color='purple', label='Densité normale')
    ax.set_title(f"n = {n}")
    ax.set_xlabel("Moyenne empirique")
    ax.set_ylabel("Densité")
    ax.legend()


plt.suptitle("Illustration du TCL avec des variables de Bernoulli", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])

plt.savefig("../rapport/figures/figure_tcl_bernoulli.png", dpi=300)

plt.show()
```
