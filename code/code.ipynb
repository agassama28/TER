```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 1000
m = 0
sigma = 1
nb_trajectoires = 5

plt.figure(figsize=(10, 6))

for i in range (nb_trajectoires) :
    X = np.random.normal(loc=m, scale=sigma, size=n)
    moyenne_empirique = np.cumsum(X) / np.arange(1, n + 1)
    x = np.linspace(0,1000,1000)
    plt.plot(x, moyenne_empirique)
plt.axhline(y= 0, color='red', linestyle='--', label="Moyenne thorique")
plt.title("Illustration de la LGN avec une loi normale centre rduite")
plt.xlabel("n")
plt.ylabel("moyenne empirique")
#plt.savefig("../figures/figure_lgn_1.png", dpi=300)
plt.show()
```

```python
a = 1
b = 2

plt.figure(figsize=(10, 6))

for i in range(nb_trajectoires) :
    X = np.random.uniform(a, b, 1000)
    moyenne_empirique = np.cumsum(X) / np.arange(1, n + 1)
    plt.plot(x, moyenne_empirique)
plt.axhline(y=(a + b) / 2, color='red', linestyle='--', label="Moyenne thorique")
plt.title("Illustration de la LGN avec une loi uniforme")
plt.xlabel("n")
plt.ylabel("moyenne empirique")
#plt.savefig("../figures/figure_lgn_2.png", dpi=300)
plt.show()
```

```python
lambda_ = 1

plt.figure(figsize=(10, 6))

for i in range (nb_trajectoires) :
    X = np.random.exponential(scale= 1 / lambda_, size=n)
    moyenne_empirique = np.cumsum(X) / np.arange(1, n + 1)
    x = np.linspace(0,1000,1000)
    plt.plot(x, moyenne_empirique)
plt.axhline(y= lambda_, color='red', linestyle='--', label="Moyenne thorique")
plt.title("Illustration de la LGN avec une loi exponentielle")
plt.xlabel("n")
plt.ylabel("moyenne empirique")
#plt.savefig("../figures/figure_lgn_3.png", dpi=300)
plt.show()
```

```python
n = 1000
nb_simulations = 1000
p = 3 / 5
moyennes = [] 

for i in range(nb_simulations) : 
    X = np.random.binomial(1, p, 1000)
    moyenne = np.mean(X)
    moyennes.append(moyenne)
plt.figure(figsize=(10, 6))
plt.hist(moyennes, bins=30, density=True, alpha=0.7, color='black')
plt.title("Illustration du TCL avec des variables de Bernoulli")
#plt.savefig("../figures/figure_tcl_1.png", dpi=300)
plt.show()
```
