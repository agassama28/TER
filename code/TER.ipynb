```python
import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
```

```python
# Initialisation des paramtres

n = 10 # nombre de sommets
p = 0.5
nb_graph = 10**4
graph_list = []

# fonction qui genere un graphe alatoire orient (i < j)

def generate_oriented_graph(n, p):
    G = nx.DiGraph()
    G.add_nodes_from(range(1, n+1))  # sommets numrots 1..n

    for i in range(1, n+1):
        for j in range(i+1, n+1):  # i < j
            if np.random.rand() < p:
                G.add_edge(i, j)
    
    return G
    
# boucle genere qui genere une liste de graphe 

for i in range(nb_graph):
    #G = nx.erdos_renyi_graph(n, p, directed= True)
    #mapping = {i: i+1 for i in G.nodes()}
    #G = nx.relabel_nodes(G, mapping)
    G = generate_oriented_graph(n, p)
    graph_list.append(G)

# on calule le ratio du nombre de graphes o 1 et n sont connects

graph_1_n = 0
for i in range(nb_graph):
    if (nx.has_path(graph_list[i], 1, n)):
        graph_1_n += 1
        
M = nx.to_numpy_array(graph_list[0]) # matrice d'adjacence
for G in graph_list:
    pos = nx.circular_layout(G)
    
nx.draw(graph_list[0], with_labels=True, pos=pos) # dessine le premier graphe
print(nx.has_path(graph_list[0], 1, n)) # affiche si 1 et n sont connectes dans le graphe numero 0
print(graph_1_n / nb_graph) #afficehe le ratio
print(M)
plt.show()
```
